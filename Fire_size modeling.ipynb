```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
# Assuming you have a DataFrame called 'fires' with the provided columns
# Features (X) and target variable (y)
X = dfs.drop('FIRE_SIZE', axis=1)
y = dfs['FIRE_SIZE']
```

```python
# Preprocess the data: drop non-numeric columns, and fill missing values
X = X.select_dtypes(include=[np.number]).fillna(0)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Initialize the models
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor()
}

# KFold Cross-Validation
k = 5  # Number of splits
kf = KFold(n_splits=k, random_state=42, shuffle=True)

# Evaluate models using k-fold cross-validation
for name, model in models.items():
    cv_scores = cross_val_score(model, X_scaled, y, cv=kf, scoring='r2')
    print(f"{name}: R2 Score Mean: {cv_scores.mean():.4f}, Std: {cv_scores.std():.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the Random Forest model
model = RandomForestRegressor()

# Define the hyperparameter search space
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Use GridSearchCV to search for the best hyperparameters
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=1)

# Fit the model with the best combination of hyperparameters
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
print("Best Parameters: ", best_params)

# Evaluate the model using the best parameters
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R2 Score: ", r2)
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import RFE, SelectFromModel
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

# Assuming you have a DataFrame called 'fires' with the provided columns
# Features (X) and target variable (y)
X = fires.drop('FIRE_SIZE', axis=1)
y = fires['FIRE_SIZE']

# Preprocess the data: drop non-numeric columns, and fill missing values
X = X.select_dtypes(include=[np.number]).fillna(0)

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Recursive Feature Elimination (RFE) with Random Forest
model_rf = RandomForestRegressor()
rfe = RFE(estimator=model_rf, n_features_to_select=5, step=1) # Adjust 'n_features_to_select' based on desired number of features
rfe.fit(X_scaled, y)
X_rfe = rfe.transform(X_scaled)
selected_features_rfe = X.columns[rfe.support_]
print("Selected Features (RFE):", selected_features_rfe)

# LASSO regularization
model_lasso = Lasso(alpha=0.1) # Adjust 'alpha' based on desired level of regularization
model_lasso.fit(X_scaled, y)
sfm = SelectFromModel(model_lasso)
sfm.fit(X_scaled, y)
X_lasso = sfm.transform(X_scaled)
selected_features_lasso = X.columns[sfm.get_support()]
print("Selected Features (LASSO):", selected_features_lasso)

# Correlation Analysis
correlations = fires.corr().abs()
threshold = 0.5  # Set a threshold for correlation
correlated_vars = correlations.loc['FIRE_SIZE'][correlations['FIRE_SIZE'] > threshold]
selected_features_corr = correlated_vars.index.tolist()
selected_features_corr.remove('FIRE_SIZE')
print("Selected Features (Correlation):", selected_features_corr)
```
